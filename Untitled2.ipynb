### **Diagrama de dispersión**
---
El diagrama de dispersión se utiliza para averiguar la intensidad de la relación entre dos variables numéricas, por esto es que las varaibles antes planteadas por variables numéricas.

En este caso, mediante el diagrama de dispersión se definira las variables explicativas y la varaible respuesta



```python
x = np.array(tabla_regresion1["ESTU_INSE_INDIVIDUAL"]).reshape(-1,1)
y = tabla_regresion1["PUNT_LECTURA_CRITICA"]
trainx, testx, trainy, testy = train_test_split(x, y, test_size = 0.2, random_state = 0)
modelo_lineal0 = LinearRegression()
regresion0 = modelo_lineal0.fit(trainx, trainy)
pred_lineal = regresion0.predict(testx)
```

```python
x0 = np.array(tabla_regresion1["ESTU_INSE_INDIVIDUAL"]).reshape(-1,1)
y0 = tabla_regresion1["PUNT_LECTURA_CRITICA"]
modelo_lineal01 = LinearRegression()
regresion01 = modelo_lineal01.fit(x0, y0)
pred_lineal01 = regresion01.predict(x0)
```

```python
plt.figure(figsize=(5, 4))
plt.scatter(tabla_regresion1["ESTU_INSE_INDIVIDUAL"], tabla_regresion1["PUNT_LECTURA_CRITICA"])
#plt.plot(testx, pred_lineal, linestyle='-', color = "red")
plt.plot(tabla_regresion1["ESTU_INSE_INDIVIDUAL"], pred_lineal01, linestyle='-', color = "green")
plt.show()
```

El anterior diagrama de dispersión se esta relacionando las variables  ESTU_INSE_INDIVIDUAL y PUNT_LECTURA_CRITICA, de acuerdo al gráfico y a la recta de regresión posiblemente estas variables presentan correlación positiva, sim embargo, esta correlación es leve, ya que el gráfico tiende a parecer a una nube de puntos sin un patron determinado

```python
plt.figure(figsize=(5, 4))
plt.scatter(tabla_regresion1["PUNT_INGLES"], tabla_regresion1["PUNT_MATEMATICAS"])
#plt.plot(testx, pred_lineal, linestyle='-', color = "red")
plt.plot(tabla_regresion1["PUNT_INGLES"], pred_lineal01, linestyle='-', color = "VIOLET")
plt.show()
```

De lo anterior se logra concluir que las variables PUNT_INGLES y PUNT_MATEMATICAS no presenta ninguna clase de correlación; esto deacuerdo a que el diagrama de dispersión es una nube de puntos sin patron especifico y no existe una recta de regresión que determine su comportamiento

```python
plt.figure(figsize=(5, 4))
plt.scatter(tabla_regresion1["PUNT_MATEMATICAS"], tabla_regresion1["PUNT_LECTURA_CRITICA"])
#plt.plot(testx, pred_lineal, linestyle='-', color = "red")
plt.plot(tabla_regresion1["PUNT_MATEMATICAS"], pred_lineal01, linestyle='-', color = "black")
plt.show()
```

Como se visualiza en el código ejecutado, el diagrama de dispersión anterior corresponde a las variables PUNT_MATEMATICAS y PUNT_LECTURA_CRITICA que son los puntos de matemáticas y lectura crítica correspondientemente de los evaluandos. Se nota que, visualmente, que no hay una correlación entre estas variables. Es decir, el puntaje de lectura crítica de los evaluados no tiene ninguna relación con el puntaje de matemáticas

### **Coeficiente de correlación de Pearson, Spearman y Kendall**



```python
coef_pearson1 = np.corrcoef(tabla_regresion1["PUNT_MATEMATICAS"], tabla_regresion1["ESTU_INSE_INDIVIDUAL"])[0, 1]
coef_spearman1, _ = spearmanr(tabla_regresion1["PUNT_MATEMATICAS"], tabla_regresion1["ESTU_INSE_INDIVIDUAL"])
coef_kendall1, _ = kendalltau(tabla_regresion1["PUNT_MATEMATICAS"], tabla_regresion1["ESTU_INSE_INDIVIDUAL"])
print(f"Los coeficientes de correlación de Pearson, Spearman y Kendall son respectivamente {coef_pearson1}, {coef_spearman1} y {coef_kendall1}.")
```

```python
coef_pearson1 = np.corrcoef(tabla_regresion1["PUNT_C_NATURALES"], tabla_regresion1["ESTU_INSE_INDIVIDUAL"])[0, 1]
coef_spearman1, _ = spearmanr(tabla_regresion1["PUNT_C_NATURALES"], tabla_regresion1["ESTU_INSE_INDIVIDUAL"])
coef_kendall1, _ = kendalltau(tabla_regresion1["PUNT_C_NATURALES"], tabla_regresion1["ESTU_INSE_INDIVIDUAL"])
print(f"Los coeficientes de correlación de Pearson, Spearman y Kendall son respectivamente {coef_pearson1}, {coef_spearman1} y {coef_kendall1}.")
```

Dadas las anteriores valores de los coeficientes, entre más alto sea el valor, esa variable será la más explicativa entre todas, en este caso lo observamos por medio de los coeficientes de correlación de Pearson, Spearman y Kendall, que la variable PUNT_INGLES es la variable más explicativa.
